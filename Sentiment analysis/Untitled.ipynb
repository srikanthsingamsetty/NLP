```python
import re
import nltk
import string
import nlp_utils 
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
with open('TextAnalytics.txt','r') as f:
    text = f.read()
```

```python
text
```

```python
text = text.split('\n')
```

```python
corpus =[text]
for sent in text:
    corpus.append(sent.split('\t'))
```

```python
letters_only=re.sub(r'[^a-zA-Z]'," ",str(corpus))
```

```python
letters_only = letters_only.lower()
token=nltk.sent_tokenize(letters_only)
token
```

```python
def num_dec_al(word):
    if word.isnumeric():
        return 'xxxxxx'
    elif word.isdecimal():
        return 'xxx...'
    elif word.isalpha():
        return word
    else:
        return'xxxaaa'

def clean_nda(token):
    token = nlp_utils.w_tokenization(token)
    map_list = list(map(num_dec_al,token))
    return " ".map_list

corpus_nda = list(map(clean_nda,token))

corpus_nda
```
